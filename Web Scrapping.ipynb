```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Open the page
driver.get("https://fbref.com/en/comps/9/stats/Premier-League-Stats")

# Implicitly wait for elements to load
driver.implicitly_wait(30)

# Find the stats table
table = driver.find_element(By.CSS_SELECTOR, "table#stats_standard")

# Extract rows
rows = table.find_elements(By.TAG_NAME, "tr")

# Loop through each row and extract text
for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")  # Get all <td> elements
    row_data = [cell.text for cell in cells]  # Extract text from each <td>
    
    if row_data:  # Ignore empty rows
        print(row_data)

# Close the driver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

# Set up Selenium WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Open the page
driver.get("https://fbref.com/en/comps/8/2023-2024/stats/2023-2024-Champions-League-Stats")

# Implicitly wait for elements to load
driver.implicitly_wait(30)

# Find the stats table
table = driver.find_element(By.CSS_SELECTOR, "table#stats_standard")

# Extract column headers (skipping first empty header if present)
headers = [th.text.strip() for th in table.find_elements(By.TAG_NAME, "th") if th.text.strip()]

# Extract rows
rows = table.find_elements(By.TAG_NAME, "tr")

# Store data in a list
data = []
for row in rows[1:]:  # Skip the header row
    cells = row.find_elements(By.TAG_NAME, "td")
    row_data = [cell.text.strip() for cell in cells]

    # Ensure we only append rows with the correct column count
    data.append(row_data)

# Convert to DataFrame
df = pd.DataFrame(data, columns=headers)

# Save to CSV file
df.to_csv("Premier_League_Player_Stats.csv", index=False)

print("✅ Data extracted and saved successfully!")

# Close the driver
driver.quit()
```
